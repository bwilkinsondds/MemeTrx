## Extract Metrics from a Single 5-Second Window

### Subtask:
For the first 5 seconds (1250 samples) of one chosen channel (Fp1 or Fp2) from the loaded awake file, calculate the following metrics: permutation entropy (with order=3), spectral entropy, alpha band power (8-13 Hz) using Welch's method, mean amplitude, and standard deviation. Print all calculated results and verify that all values are numbers (not NaN).


**Reasoning**:
I will calculate the required EEG metrics for the first 5-second segment of the chosen channel, including permutation entropy, spectral entropy, alpha band power, mean amplitude, and standard deviation, and then print and verify these results. I will implement a permutation entropy function as `nolds` is not guaranteed to be installed.



````python
import numpy as np
from scipy.signal import welch
from scipy.stats import entropy
import math

# 1. Select the plot_channel and extract the first 5 seconds of data
if plot_channel is None:
    print("Error: No suitable channel was identified for plotting.")
else:
    # Get the index of the selected channel
    ch_idx = raw.ch_names.index(plot_channel)

    # Extract data for the first 5 seconds (5 * sfreq samples)
    # The raw data is already downsampled to target_sfreq (250 Hz)
    n_samples_5s = int(5 * raw.info['sfreq'])
    data_segment = raw._data[ch_idx, :n_samples_5s]

    print(f"\nExtracted {len(data_segment)} samples for {plot_channel} (5 seconds at {raw.info['sfreq']} Hz).")

    # 2. Define permutation_entropy function (without nolds)
    def calculate_permutation_entropy(x, order=3, delay=1):
        if len(x) < order + (order - 1) * delay:
            return np.nan # Not enough data points

        permutations = []
        for i in range(len(x) - (order - 1) * delay):
            # Create an ordinal pattern
            pattern = x[i : i + order * delay : delay]
            sorted_indices = np.argsort(pattern)
            permutations.append(tuple(sorted_indices))

        # Count occurrences of each permutation
        unique_permutations, counts = np.unique(permutations, return_counts=True, axis=0)

        # Calculate probabilities
        probabilities = counts / len(permutations)

        # Calculate Shannon entropy
        # Normalize by log2(factorial(order)) for a normalized result between 0 and 1 if normalize=True
        # The request does not explicitly ask for normalization, so we will return the direct Shannon entropy
        pe = -np.sum(probabilities * np.log2(probabilities))
        return pe

    permutation_ent = calculate_permutation_entropy(data_segment, order=3)
    print(f"Permutation Entropy (order=3): {permutation_ent:.4f}")

    # 3. Calculate Power Spectral Density (PSD) using Welch's method
    sfreq = raw.info['sfreq']
    freqs, psd = welch(data_segment, sfreq, nperseg=int(2 * sfreq)) # 2-second windows for Welch

    # 4. Compute Spectral Entropy
    # Normalize PSD to form a probability distribution
    psd_norm = psd / np.sum(psd)
    # Avoid log2(0) by replacing 0s with a very small number
    psd_norm[psd_norm == 0] = np.finfo(float).eps
    spectral_ent = entropy(psd_norm, base=2) # Using base 2 for bits
    print(f"Spectral Entropy: {spectral_ent:.4f}")

    # 5. Compute Alpha Band Power (8-13 Hz)
    idx_alpha = np.logical_and(freqs >= 8, freqs <= 13)
    alpha_power = np.trapz(psd[idx_alpha], freqs[idx_alpha])
    print(f"Alpha Band Power (8-13 Hz): {alpha_power:.4f}")

    # 6. Calculate Mean Amplitude
    mean_amplitude = np.mean(data_segment)
    print(f"Mean Amplitude: {mean_amplitude:.4f}")

    # 7. Calculate Standard Deviation
    std_dev = np.std(data_segment)
    print(f"Standard Deviation: {std_dev:.4f}")

    # 8. Print all calculated results (already done above)

    # 9. Verify that none of the calculated metrics are NaN
    metrics = {
        "Permutation Entropy": permutation_ent,
        "Spectral Entropy": spectral_ent,
        "Alpha Band Power": alpha_power,
        "Mean Amplitude": mean_amplitude,
        "Standard Deviation": std_dev
    }

    print("\n--- NaN Verification ---")
    all_numeric = True
    for name, value in metrics.items():
        if np.isnan(value):
            print(f"Warning: {name} is NaN!")
            all_numeric = False
        else:
            print(f"{name}: {value} (Numeric)")

    if all_numeric:
        print("All calculated metrics are numeric (not NaN).")
    else:
        print("Some calculated metrics are NaN. Please check the data or calculations.")
````

## Process All Windows for Awake State

### Subtask:
Process the awake EEG data using 5-second windows with a 1-second step size. For each window, extract the same metrics as in the previous step (permutation entropy, spectral entropy, alpha power, mean amplitude, standard deviation) using only one channel (Fp1 or Fp2). Create a pandas DataFrame with columns: `time_sec`, `perm_entropy`, `spectral_entropy`, `alpha_power`, `mean_amp`, `std_amp`. Display the first 10 rows of this DataFrame and use the `describe()` method to show summary statistics.


**Reasoning**:
I will process the awake EEG data in 5-second windows with a 1-second step size, calculate the specified metrics for each window, and store the results in a pandas DataFrame. Finally, I will display the first 10 rows and summary statistics of this DataFrame.



````python
import pandas as pd

# Initialize an empty list to store metrics for each window
all_metrics_awake = []

# Determine the channel index for the chosen plotting channel (Fp1 or Fp2)
# 'plot_channel' was determined in a previous step
if plot_channel is None:
    print("Error: No suitable channel was identified. Cannot proceed.")
else:
    ch_idx = raw.ch_names.index(plot_channel)
    sfreq = raw.info['sfreq'] # Sampling frequency (already 250 Hz)

    # Define window and step sizes in samples
    window_size_sec = 5
    step_size_sec = 1
    window_size_samples = int(window_size_sec * sfreq)
    step_size_samples = int(step_size_sec * sfreq)

    print(f"Processing awake data for channel: {plot_channel} with window_size={window_size_sec}s and step_size={step_size_sec}s...")

    # Iterate through the EEG data with sliding windows
    # Ensure the window fully fits within the data
    total_samples = raw._data.shape[1]
    for start_sample in range(0, total_samples - window_size_samples + 1, step_size_samples):
        end_sample = start_sample + window_size_samples
        data_segment = raw._data[ch_idx, start_sample:end_sample]

        # Calculate time in seconds for the start of the window
        time_sec = start_sample / sfreq

        # Calculate metrics for the current window's data_segment
        # Permutation Entropy
        perm_entropy = calculate_permutation_entropy(data_segment, order=3)

        # Power Spectral Density (PSD) using Welch's method
        freqs, psd = welch(data_segment, sfreq, nperseg=int(2 * sfreq)) # 2-second windows for Welch

        # Spectral Entropy
        psd_norm = psd / np.sum(psd)
        psd_norm[psd_norm == 0] = np.finfo(float).eps # Avoid log2(0)
        spectral_entropy = entropy(psd_norm, base=2)

        # Alpha Band Power (8-13 Hz)
        idx_alpha = np.logical_and(freqs >= 8, freqs <= 13)
        alpha_power = trapezoid(psd[idx_alpha], freqs[idx_alpha])

        # Mean Amplitude
        mean_amp = np.mean(data_segment)

        # Standard Deviation
        std_amp = np.std(data_segment)

        # Store metrics in a dictionary
        metrics_dict = {
            "time_sec": time_sec,
            "perm_entropy": perm_entropy,
            "spectral_entropy": spectral_entropy,
            "alpha_power": alpha_power,
            "mean_amp": mean_amp,
            "std_amp": std_amp,
            "state": "Awake" # Add state for later comparison
        }
        all_metrics_awake.append(metrics_dict)

    # Convert the list of dictionaries to a pandas DataFrame
    df_awake_metrics = pd.DataFrame(all_metrics_awake)

    print("\n--- First 10 rows of Awake Metrics DataFrame ---")
    print(df_awake_metrics.head(10))

    print("\n--- Summary Statistics of Awake Metrics DataFrame ---")
    print(df_awake_metrics.describe())
````

## Process Other States and Combine Data

### Subtask:
Repeat the window-based metric extraction process (as described in the previous step) for the 'sedation' and 'recovery' EEG files, if they were identified and exist. If a state file does not exist, continue processing with the states that are available. Combine the resulting DataFrames from all processed states (awake, sedation, recovery) into a single master DataFrame, adding a new 'state' column to identify the origin of each row. Display the counts of entries per state to ensure correct combination.


**Reasoning**:
I will define a function to abstract the metric extraction process for a given EEG file and state. Then, I will apply this function to the sedation and recovery files (if they exist) and concatenate all state-specific DataFrames into a single combined DataFrame.
````python
import pandas as pd
import mne
import numpy as np
from scipy.signal import welch
from scipy.stats import entropy
from scipy.integrate import trapezoid

# Re-define the helper function for permutation entropy in case it's not in scope
def calculate_permutation_entropy(x, order=3, delay=1):
    if len(x) < order + (order - 1) * delay:
        return np.nan # Not enough data points

    permutations = []
    for i in range(len(x) - (order - 1) * delay):
        pattern = x[i : i + order * delay : delay]
        sorted_indices = np.argsort(pattern)
        permutations.append(tuple(sorted_indices))

    if not permutations:
        return np.nan

    unique_permutations, counts = np.unique(permutations, return_counts=True, axis=0)
    probabilities = counts / len(permutations)
    pe = -np.sum(probabilities * np.log2(probabilities))
    return pe

def process_eeg_file(file_path, state_name, target_sfreq=250):
    if not file_path or not os.path.exists(file_path):
        print(f"Skipping {state_name}: File path is invalid or does not exist: {file_path}")
        return pd.DataFrame() # Return empty DataFrame if file not found

    print(f"\nProcessing {state_name} EEG file: {file_path}")
    raw_current = mne.io.read_raw_brainvision(file_path, preload=True, verbose=False)

    # Downsample if necessary
    if raw_current.info['sfreq'] != target_sfreq:
        print(f"Downsampling {state_name} data from {raw_current.info['sfreq']} Hz to {target_sfreq} Hz...")
        raw_current.resample(sfreq=target_sfreq, npad="auto", verbose=False)
    else:
        print(f"{state_name} data already at {target_sfreq} Hz.")

    sfreq = raw_current.info['sfreq']

    # Identify channel for processing (Fp1 or Fp2, or first EEG channel)
    process_channel = None
    if 'Fp1' in raw_current.ch_names:
        process_channel = 'Fp1'
    elif 'Fp2' in raw_current.ch_names:
        process_channel = 'Fp2'
    else:
        eeg_channels = mne.pick_types(raw_current.info, eeg=True, exclude='bads')
        if len(eeg_channels) > 0:
            process_channel = raw_current.ch_names[eeg_channels[0]]

    if process_channel is None:
        print(f"Error: No suitable EEG channel found for {state_name}. Skipping.")
        return pd.DataFrame()

    ch_idx = raw_current.ch_names.index(process_channel)
    print(f"Using channel '{process_channel}' for {state_name} analysis.")

    all_metrics = []
    window_size_sec = 5
    step_size_sec = 1
    window_size_samples = int(window_size_sec * sfreq)
    step_size_samples = int(step_size_sec * sfreq)

    total_samples = raw_current._data.shape[1]

    for start_sample in range(0, total_samples - window_size_samples + 1, step_size_samples):
        end_sample = start_sample + window_size_samples
        data_segment = raw_current._data[ch_idx, start_sample:end_sample]

        time_sec = start_sample / sfreq

        perm_entropy = calculate_permutation_entropy(data_segment, order=3)

        freqs, psd = welch(data_segment, sfreq, nperseg=int(2 * sfreq))

        psd_norm = psd / np.sum(psd)
        psd_norm[psd_norm == 0] = np.finfo(float).eps
        spectral_entropy = entropy(psd_norm, base=2)

        idx_alpha = np.logical_and(freqs >= 8, freqs <= 13)
        alpha_power = trapezoid(psd[idx_alpha], freqs[idx_alpha])

        mean_amp = np.mean(data_segment)
        std_amp = np.std(data_segment)

        metrics_dict = {
            "time_sec": time_sec,
            "perm_entropy": perm_entropy,
            "spectral_entropy": spectral_entropy,
            "alpha_power": alpha_power,
            "mean_amp": mean_amp,
            "std_amp": std_amp,
            "state": state_name
        }
        all_metrics.append(metrics_dict)

    return pd.DataFrame(all_metrics)

# Process Sedation state
df_sedation_metrics = process_eeg_file(sedation_file, "Sedation")

# Process Recovery state
df_recovery_metrics = process_eeg_file(recovery_file, "Recovery")

# Combine all DataFrames
all_dfs = [df_awake_metrics]
if not df_sedation_metrics.empty:
    all_dfs.append(df_sedation_metrics)
if not df_recovery_metrics.empty:
    all_dfs.append(df_recovery_metrics)

df_combined_metrics = pd.concat(all_dfs, ignore_index=True)

print("\n--- Counts of entries per state in Combined DataFrame ---")
print(df_combined_metrics['state'].value_counts())
````

````python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# --- TEST MODE CONFIGURATION ---
TEST_MODE = True  # Set to False to run full analysis after testing
TEST_WINDOWS_LIMIT = 20 # Number of windows to process in TEST_MODE
MAX_BETTI_POINTS = 500 # Max number of points for Betti calculation (subsamples if exceeded)
# --- END TEST MODE CONFIGURATION ---

# ============================================================================
# STEP 1: LOAD YOUR EXISTING DATA
# ============================================================================
print("=" * 60)
print("STEP 1: Loading existing EEG data...")
print("=" * 60)

df = pd.read_csv('/content/eeg_metrics_complete.csv')
print(f"Loaded {len(df)} windows: {df['state'].value_counts().to_dict()}")

# ============================================================================
# STEP 2: FIXED COMPLEXITY FUNCTIONS (NO MORE NaN)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 2: Defining fixed complexity metrics (excluding Sample Entropy)...")
print("=" * 60)

# Sample Entropy function removed as per user request

def lempel_ziv_complexity_fixed(signal_data):
    """Fixed Lempel-Ziv Complexity"""
    if len(np.unique(signal_data)) == 1:
        return 0.0

    threshold = np.median(signal_data)
    binary = (signal_data > threshold).astype(int)

    if np.all(binary == 0) or np.all(binary == 1):
        threshold = np.mean(signal_data)
        binary = (signal_data > threshold).astype(int)
        if np.all(binary == 0) or np.all(binary == 1):
            return 0.0

    binary_str = ''.join(binary.astype(str))
    n = len(binary_str)
    complexity = 1
    prefix_len = 1
    i = 0

    while i < n:
        prefix = binary_str[i:i+prefix_len]
        subsequence = binary_str[:i]
        if prefix in subsequence:
            prefix_len += 1
            if i + prefix_len > n:
                break
        else:
            complexity += 1
            i += prefix_len
            prefix_len = 1

    if n > 1:
        max_complexity = n / np.log2(n)
        normalized = complexity / max_complexity
    else:
        normalized = 0.0

    return normalized

print(" Fixed complexity functions loaded (Lempel-Ziv only)")

# ============================================================================
# STEP 3: TOPOLOGY FUNCTION (BETTI NUMBERS)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 3: Defining topology metrics (Betti numbers)...")
print("=" * 60)

def compute_betti_numbers_simple(signal_data, embed_dim=3, tau=10, epsilon_percentile=10, max_points_limit=MAX_BETTI_POINTS):
    """
    Compute Betti numbers from time-delay embedding

    Parameters:
    - signal_data: 1D array of EEG signal
    - embed_dim: embedding dimension (3 for visualization)
    - tau: time delay
    - epsilon_percentile: percentile of distances to use as epsilon
    - max_points_limit: maximum number of embedded points to process. Subsamples if exceeded.

    Returns:
    - beta_0: number of connected components
    - beta_1: number of 1D holes (loops)
    """

    # Time-delay embedding
    n = len(signal_data)
    n_points_raw = n - (embed_dim - 1) * tau

    if n_points_raw < 10:
        return np.nan, np.nan

    embedded_raw = np.zeros((n_points_raw, embed_dim))
    for i in range(n_points_raw):
        for j in range(embed_dim):
            embedded_raw[i, j] = signal_data[i + j * tau]

    # Subsampling if n_points exceeds limit
    if n_points_raw > max_points_limit:
        print(f"    Subsampling {n_points_raw} embedded points to {max_points_limit} for Betti number calculation.")
        idx = np.random.choice(n_points_raw, max_points_limit, replace=False)
        embedded = embedded_raw[idx]
        n_points = max_points_limit
    else:
        embedded = embedded_raw
        n_points = n_points_raw

    # Compute pairwise distances
    distances = pdist(embedded, metric='euclidean')
    dist_matrix = squareform(distances)

    # Use percentile as epsilon (filtration parameter)
    epsilon = np.percentile(distances, epsilon_percentile)

    # Beta_0: Connected components using DBSCAN
    clustering = DBSCAN(eps=epsilon, min_samples=2).fit(embedded)
    beta_0 = len(set(clustering.labels_)) - (1 if -1 in clustering.labels_ else 0)

    # Beta_1: Simple approximation - count triangles that form loops
    # This is a simplified version; full persistent homology is more complex
    adjacency = (dist_matrix < epsilon).astype(int)
    np.fill_diagonal(adjacency, 0)

    # Count 3-cycles (rough proxy for beta_1)
    adjacency_squared = adjacency @ adjacency
    adjacency_cubed = adjacency_squared @ adjacency
    n_triangles = np.trace(adjacency_cubed) // 6  # Each triangle counted 6 times

    # Beta_1 approximation: independent cycles
    # This is crude but gives signal
    beta_1 = max(0, n_triangles - beta_0 + 1)  # Euler characteristic approximation

    return beta_0, beta_1

print(" Betti number function loaded")

# ============================================================================
# STEP 4: RE-PROCESS DATA WITH NEW METRICS
# ============================================================================
print("\n" + "=" * 60)
print("STEP 4: Re-loading raw EEG and computing new metrics...")
print("=" * 60)

import mne

# Load raw data
awake_file = !ls /content/ds005620/sub-1010/eeg/*awake*.vhdr
sed_file = !ls /content/ds005620/sub-1010/eeg/*sed_run*.vhdr

if not awake_file:
    awake_file = !ls /content/ds005620/sub-1010/eeg/*task-awake*.vhdr

results = []

states_to_process = [('Awake', awake_file[0])]
if not TEST_MODE:
    states_to_process.append(('Sedation', sed_file[0]))

for state, filepath in states_to_process:
    print(f"\nProcessing {state} state...")

    raw = mne.io.read_raw_brainvision(filepath, preload=True, verbose=False)
    raw.resample(250, verbose=False)

    channel = 'Fp1' if 'Fp1' in raw.ch_names else raw.ch_names[0]
    data = raw.get_data(picks=[channel])[0]

    sfreq = raw.info['sfreq']
    window_samples = int(5.0 * sfreq)
    step_samples = int(1.0 * sfreq)

    n_windows_state = (len(data) - window_samples) // step_samples + 1

    if TEST_MODE:
        n_windows_state = min(n_windows_state, TEST_WINDOWS_LIMIT)

    for i in range(n_windows_state):
        if i % 10 == 0:
            print(f"  Processing window {i}/{n_windows_state} for {state} state...")

        start = i * step_samples
        end = start + window_samples
        window = data[start:end]

        # Original metrics
        perm_ent = df[(df['state'] == state) & (df.index == i)]['perm_entropy'].values[0] if i < len(df[df['state'] == state]) else np.nan
        spec_ent = df[(df['state'] == state) & (df.index == i)]['spectral_entropy'].values[0] if i < len(df[df['state'] == state]) else np.nan
        alpha_pow = df[(df['state'] == state) & (df.index == i)]['alpha_power'].values[0] if i < len(df[df['state'] == state]) else np.nan

        # Lempel-Ziv Complexity
        lz_comp = lempel_ziv_complexity_fixed(window)

        # Topology metrics
        beta_0, beta_1 = compute_betti_numbers_simple(window)

        results.append({
            'state': state,
            'time_sec': i,
            'perm_entropy': perm_ent,
            'spectral_entropy': spec_ent,
            'alpha_power': alpha_pow,
            'lempel_ziv_complexity': lz_comp,
            'betti_0': beta_0,
            'betti_1': beta_1
        })

    print(f"  Finished processing {n_windows_state} windows for {state} state.")

df_enhanced = pd.DataFrame(results)

if TEST_MODE:
    print("\n--- TEST MODE RESULTS ---")
    test_beta_1 = df_enhanced['betti_1']
    print(f"Mean Beta_1: {test_beta_1.mean():.4f}")
    print(f"Number of non-NaN Beta_1 values: {test_beta_1.count()}")
    print(f"All Beta_1 values are numeric: {test_beta_1.notna().all()}")
    print("-------------------------")
else:

    print("\n" + "=" * 60)
    print("STEP 5: RESULTS SUMMARY")
    print("=" * 60)

    print("\nDataset Overview:")
    print(df_enhanced.groupby('state').size())

    print("\nMetrics Summary by State:")
    print(df_enhanced.groupby('state')[['perm_entropy',
                                         'lempel_ziv_complexity', 'alpha_power',
                                         'betti_0', 'betti_1']].mean())

    # ============================================================================
    # STEP 6: STATISTICAL TESTS
    # ============================================================================
    print("\n" + "=" * 60)
    print("STEP 6: Statistical Comparisons (ANOVA)")
    print("=" * 60)

    for metric in ['perm_entropy', 'lempel_ziv_complexity',
                   'alpha_power', 'betti_0', 'betti_1']:
        awake_vals = df_enhanced[df_enhanced['state'] == 'Awake'][metric].dropna()
        sed_vals = df_enhanced[df_enhanced['state'] == 'Sedation'][metric].dropna()

        if len(awake_vals) > 0 and len(sed_vals) > 0:
            f_stat, p_val = stats.f_oneway(awake_vals, sed_vals)
            print(f"{metric:25s}: F={f_stat:8.2f}, p={p_val:.4f}")

    # ============================================================================
    # STEP 7: VISUALIZATIONS
    # ============================================================================
    print("\n" + "=" * 60)
    print("STEP 7: Creating visualizations...")
    print("=" * 60)

    # We now have 5 metrics: perm_entropy, lempel_ziv_complexity, alpha_power, betti_0, betti_1
    # Adjusting subplot layout to 2 rows, 3 columns to accommodate 5 plots nicely
    fig, axes = plt.subplots(2, 3, figsize=(18, 12)) # Increased figure size for better readability
    axes = axes.flatten()

    metrics_to_plot = ['perm_entropy', 'lempel_ziv_complexity',
                       'alpha_power', 'betti_0', 'betti_1']

    for idx, metric in enumerate(metrics_to_plot):
        sns.boxplot(x='state', y=metric, data=df_enhanced, ax=axes[idx])
        axes[idx].set_title(f'{metric.replace("_", " ").title()}')
        axes[idx].set_xlabel('State')
        axes[idx].set_ylabel(metric.replace("_", " ").title())

    # Hide any unused subplots if the total number of plots is less than the grid size
    for i in range(len(metrics_to_plot), len(axes)):
        fig.delaxes(axes[i])

    plt.suptitle('Self-Plex Metrics: Awake vs Sedation', fontsize=18, y=1.02) # Adjust suptitle position
    plt.tight_layout()
    plt.savefig('self_plex_complete_analysis.png', dpi=300, bbox_inches='tight')
    print(" Saved: self_plex_complete_analysis.png")

    # ============================================================================
    # STEP 8: SAVE RESULTS
    # ============================================================================
    print("\n" + "=" * 60)
    print("STEP 8: Saving enhanced dataset...")
    print("=" * 60)

    df_enhanced.to_csv('/content/eeg_metrics_ENHANCED.csv', index=False)
    print(" Saved: /content/eeg_metrics_ENHANCED.csv")

    # ============================================================================
    # STEP 9: SELF-PLEX VALIDATION SUMMARY
    # ============================================================================
    print("\n" + "=" * 60)
    print("SELF-PLEX FRAMEWORK VALIDATION RESULTS")
    print("=" * 60)

    awake_stats = df_enhanced[df_enhanced['state'] == 'Awake'][metrics_to_plot].mean()
    sed_stats = df_enhanced[df_enhanced['state'] == 'Sedation'][metrics_to_plot].mean()

    print("\nPREDICTED CHANGES (Self-Plex Framework):")
    print("  Entropy \u2192 DECREASE during sedation")
    print("  Alpha Power \u2192 INCREASE during sedation")
    print("  Betti-1 (loops) \u2192 DECREASE during sedation")

    print("\nOBSERVED CHANGES:")
    for metric in metrics_to_plot:
        awake_val = awake_stats[metric]
        sed_val = sed_stats[metric]
        # Handle cases where awake_val might be zero to prevent division by zero
        if pd.isna(awake_val) or awake_val == 0:
            change = np.nan # or handle as appropriate if 0 is expected for some metrics
            direction = "N/A"
        else:
            change = ((sed_val - awake_val) / awake_val * 100)
            direction = "\u2191" if change > 0 else ("\u2193" if change < 0 else "\u2192")

        print(f"  {metric:25s}: {awake_val:8.4f} \u2192 {sed_val:8.4f} ({change:+6.1f}% {direction})")

    print("\n" + "=" * 60)
    print("COMPLETE! Download the files:")
    print("  - eeg_metrics_ENHANCED.csv")
    print("  - self_plex_complete_analysis.png")
    print("=" * 60)
````

````python
import numpy as np
from scipy.spatial.distance import pdist
from sklearn.cluster import DBSCAN
import gudhi as gd # Import Gudhi

# ============================================================================
# STEP 3: TOPOLOGY FUNCTION (BETTI NUMBERS) - REFACTORED WITH GUDHI
# ============================================================================
print("\n" + "=" * 60)
print("STEP 3: Defining topology metrics (Betti numbers) with Gudhi...")
print("=" * 60)

def compute_betti_numbers_gudhi(signal_data, embed_dim=3, tau=10, epsilon_percentile=10, persistence_threshold=0.1, max_points_limit=500):
    """
    Compute Betti numbers from time-delay embedding using Gudhi for Beta_1.

    Parameters:
    - signal_data: 1D array of EEG signal
    - embed_dim: embedding dimension
    - tau: time delay
    - epsilon_percentile: percentile of distances to use as epsilon for DBSCAN (Beta_0)
    - persistence_threshold: minimum persistence for Beta_1 features to be counted
    - max_points_limit: maximum number of embedded points to process. Subsamples if exceeded.

    Returns:
    - beta_0: number of connected components (DBSCAN based)
    - beta_1_filtered: number of 1D holes (loops) filtered by persistence (Gudhi based)
    - total_beta_1: total number of 1D holes before filtering
    - top5_beta_1_lifetimes: list of top 5 longest-lived Beta_1 features
    """

    # Time-delay embedding
    n = len(signal_data)
    n_points_raw = n - (embed_dim - 1) * tau

    if n_points_raw < embed_dim + 1: # Need at least embed_dim + 1 points for a simplex
        return np.nan, np.nan, np.nan, []

    embedded_raw = np.zeros((n_points_raw, embed_dim))
    for i in range(n_points_raw):
        for j in range(embed_dim):
            embedded_raw[i, j] = signal_data[i + j * tau]

    # Subsampling if n_points exceeds limit
    if n_points_raw > max_points_limit:
        print(f"      Subsampling {n_points_raw} embedded points to {max_points_limit} for Betti number calculation.")
        idx = np.random.choice(n_points_raw, max_points_limit, replace=False)
        embedded_points = embedded_raw[idx]
        n_points = max_points_limit
    else:
        embedded_points = embedded_raw
        n_points = n_points_raw

    if n_points < embed_dim + 1: # After subsampling, check again
        return np.nan, np.nan, np.nan, []

    # Beta_0: Connected components using DBSCAN (as before)
    # Compute pairwise distances for epsilon if not already done
    distances = pdist(embedded_points, metric='euclidean')
    epsilon = np.percentile(distances, epsilon_percentile) if len(distances) > 0 else 0.0

    if epsilon > 0:
        clustering = DBSCAN(eps=epsilon, min_samples=2).fit(embedded_points)
        beta_0 = len(set(clustering.labels_)) - (1 if -1 in clustering.labels_ else 0)
    else:
        beta_0 = 1 if len(embedded_points) > 0 else 0 # If epsilon is 0, all points are their own cluster unless min_samples satisfied

    # Beta_1: Using Gudhi for persistent homology
    # Create a Rips complex from the point cloud
    # The maximum epsilon for RipsComplex should be chosen carefully.
    # For simplicity, we can use the max distance or a very large value to ensure all connections are considered for homology.
    # Using a max_edge_length that encompasses a good range of distances, for example, the 90th percentile of distances.
    if len(distances) > 0: # Ensure distances are not empty
        max_rips_epsilon = np.percentile(distances, 90) # Adjust as needed
    else:
        max_rips_epsilon = 1.0 # Default if no distances

    rips_complex = gd.RipsComplex(points=embedded_points, max_edge_length=max_rips_epsilon)
    simplex_tree = rips_complex.create_simplex_tree(max_dimension=2) # Max dimension 2 to get H1 (Beta_1)

    # Compute persistence diagrams
    persistence = simplex_tree.persistence(min_persistence=0.0) # Calculate all persistence pairs

    # Extract Beta_1 features (dimension 1)
    beta_1_features = [(birth, death) for dim, (birth, death) in persistence if dim == 1]
    total_beta_1 = len(beta_1_features)

    # Filter Beta_1 features by persistence threshold
    beta_1_filtered_features = [f for f in beta_1_features if (f[1] - f[0]) > persistence_threshold]
    beta_1_filtered = len(beta_1_filtered_features)

    # Get top 5 longest-lived Beta_1 features
    if beta_1_filtered_features:
        beta_1_lifetimes = sorted([(death - birth) for birth, death in beta_1_filtered_features], reverse=True)
        top5_beta_1_lifetimes = beta_1_lifetimes[:5]
    else:
        top5_beta_1_lifetimes = []

    return beta_0, beta_1_filtered, total_beta_1, top5_beta_1_lifetimes

print(" Refactored Betti number function with Gudhi loaded")

# ============================================================================
# STEP 4: RE-PROCESS DATA WITH NEW METRICS (Using the refactored function)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 4: Re-loading raw EEG and computing new metrics (Gudhi enabled)...")
print("=" * 60)

import mne

# Load raw data
awake_file = !ls /content/ds005620/sub-1010/eeg/*awake*.vhdr
# Ensure we pick the EC file if both are present, as used previously
awake_file_path = [f for f in awake_file if 'acq-EC' in f][0] if awake_file else None

sed_file = !ls /content/ds005620/sub-1010/eeg/*sed_run*.vhdr
# Ensure we pick run-1, as used previously
sed_file_path = [f for f in sed_file if 'run-1' in f][0] if sed_file else None

results = []

# TEST_MODE CONFIGURATION (from original notebook cell)
TEST_MODE = True  # Set to False to run full analysis after testing
TEST_WINDOWS_LIMIT = 20 # Number of windows to process in TEST_MODE
MAX_BETTI_POINTS = 500 # Max number of points for Betti calculation (subsamples if exceeded)

states_to_process = [('Awake', awake_file_path)]
if not TEST_MODE:
    if sed_file_path: # Only add sedation if file exists
        states_to_process.append(('Sedation', sed_file_path))

for state, filepath in states_to_process:
    if not filepath or not os.path.exists(filepath):
        print(f"Skipping {state}: File path is invalid or does not exist: {filepath}")
        continue

    print(f"\nProcessing {state} state...")

    raw = mne.io.read_raw_brainvision(filepath, preload=True, verbose=False)
    raw.resample(250, verbose=False)

    channel = 'Fp1' if 'Fp1' in raw.ch_names else raw.ch_names[0] # Using Fp1 or first channel
    data = raw.get_data(picks=[channel])[0]

    sfreq = raw.info['sfreq']
    window_samples = int(5.0 * sfreq)
    step_samples = int(1.0 * sfreq)

    n_windows_state = (len(data) - window_samples) // step_samples + 1

    if TEST_MODE:
        n_windows_state = min(n_windows_state, TEST_WINDOWS_LIMIT)

    # Special handling for single window test
    if TEST_MODE and TEST_WINDOWS_LIMIT == 1:
        print("--- Running single window test ---")

        start = 0 * step_samples
        end = start + window_samples
        window = data[start:end]

        beta_0, beta_1_filtered, total_beta_1, top5_beta_1_lifetimes = compute_betti_numbers_gudhi(window)

        print(f"\nResults for the first 'Awake' window:")
        print(f"  Total Beta_1 features: {total_beta_1}")
        print(f"  Filtered Beta_1 features (persistence > {0.1}): {beta_1_filtered}")
        print(f"  Top 5 longest-lived filtered Beta_1 features (lifetimes): {top5_beta_1_lifetimes}")

        # Add this single window result to results list for consistency if TEST_WINDOWS_LIMIT is 1
        # For this specific task, we're asked to process *all* 20 windows later, so skip adding this single result
        # for the final df_enhanced if TEST_WINDOWS_LIMIT is 1 for testing purposes.
        # Instead, we will reset results and run the loop for 20 windows.
        results = [] # Clear previous single window test result to re-run for 20 windows

    # Run for the specified number of windows (TEST_WINDOWS_LIMIT for TEST_MODE)
    for i in range(n_windows_state):
        if i % 10 == 0: # Print progress every 10 windows
            print(f"  Processing window {i}/{n_windows_state} for {state} state...")

        start = i * step_samples
        end = start + window_samples
        window = data[start:end]

        # Original metrics (retrieved from existing df if available, else NaN)
        # Note: This assumes `df` correctly maps index `i` to time `i` for each state.
        # For TEST_MODE, it's safer to recalculate or ensure 'df' was created with these exact metrics.
        # Given the previous context, `df` stores metrics already. Let's retrieve them if possible.
        current_df_state = df[df['state'] == state]

        perm_ent = current_df_state.iloc[i]['perm_entropy'] if i < len(current_df_state) else np.nan
        spec_ent = current_df_state.iloc[i]['spectral_entropy'] if i < len(current_df_state) else np.nan
        alpha_pow = current_df_state.iloc[i]['alpha_power'] if i < len(current_df_state) else np.nan

        # Lempel-Ziv Complexity
        lz_comp = lempel_ziv_complexity_fixed(window)

        # Topology metrics (Refactored)
        beta_0, beta_1_filtered, total_beta_1, top5_beta_1_lifetimes = compute_betti_numbers_gudhi(window,
                                                                                                   max_points_limit=MAX_BETTI_POINTS)

        results.append({
            'state': state,
            'time_sec': i,
            'perm_entropy': perm_ent,
            'spectral_entropy': spec_ent,
            'alpha_power': alpha_pow,
            'lempel_ziv_complexity': lz_comp,
            'betti_0': beta_0,
            'betti_1_filtered': beta_1_filtered, # New filtered Beta_1
            'total_beta_1': total_beta_1 # New total Beta_1 (for verification)
        })

    print(f"  Finished processing {n_windows_state} windows for {state} state.")

df_enhanced = pd.DataFrame(results)

if TEST_MODE:
    print("\n--- TEST MODE RESULTS (Gudhi Beta_1) ---")
    print(f"Processed {len(df_enhanced)} windows in TEST_MODE.")

    # Test for the first window (as explicitly requested)
    if not df_enhanced.empty:
        first_window_data = df_enhanced.iloc[0]
        print(f"\nResults for the first 'Awake' window:")
        print(f"  Total Beta_1 features: {first_window_data['total_beta_1']}")
        print(f"  Filtered Beta_1 features (persistence > {0.1}): {first_window_data['betti_1_filtered']}")
        # To get top 5 longest-lived features for the first window, we'd need to re-run the `compute_betti_numbers_gudhi`
        # function specifically for that window with the return value for top5_beta_1_lifetimes, as it's not stored in the DataFrame.
        # For now, print the lifetimes for the first window (re-calculating for clarity as it's a specific request).
        first_window_raw_data = raw.get_data(picks=[channel])[0][0*step_samples : 0*step_samples + window_samples]
        _, _, _, top5_lifetimes = compute_betti_numbers_gudhi(first_window_raw_data, max_points_limit=MAX_BETTI_POINTS)
        print(f"  Top 5 longest-lived filtered Beta_1 features (lifetimes): {top5_lifetimes}")

    # Calculate mean filtered beta_1 for the processed windows
    mean_filtered_beta_1 = df_enhanced['betti_1_filtered'].mean()
    print(f"\nMean filtered Beta_1 over {len(df_enhanced)} test windows: {mean_filtered_beta_1:.4f}")
    print("-----------------------------------------")
else:
    # (Original full analysis code for non-TEST_MODE would go here)
    print("Full analysis mode not implemented in this step.")
````

## Refine Topology Function with GTDA

### Subtask:
Modify the topology metric calculation to use `gtda`'s `VietorisRipsPersistence` and `PersistenceEntropy`. This will involve creating a `VietorisRipsPersistence` object with specified homology dimensions (e.g., [0, 1, 2] as suggested), transforming the embedded data to generate persistence diagrams, and then using `PersistenceEntropy` to quantify the complexity of these diagrams. This will replace the previous Betti number approximation and `gudhi` based Betti number calculation.


**Reasoning**:
The subtask requires refactoring the topology metric calculation to use `giotto-tda`'s `VietorisRipsPersistence` and `PersistenceEntropy`. I will modify the main processing loop by defining a new helper function `compute_gtda_metrics` that encapsulates the time-delay embedding, `VietorisRipsPersistence` for diagram generation, and `PersistenceEntropy` for computing the new metric. This function will be called in the loop, and its output will replace the previous Betti numbers.



````python
import numpy as np
import pandas as pd
import mne
import os

# Import gtda components
from gtda.homology import VietorisRipsPersistence
from gtda.diagrams import PersistenceEntropy

# ============================================================================
# HELPER FUNCTION FOR Lempel-Ziv Complexity (kept from previous steps)
# ============================================================================
def lempel_ziv_complexity_fixed(signal_data):
    """Fixed Lempel-Ziv Complexity"""
    if len(np.unique(signal_data)) == 1:
        return 0.0

    threshold = np.median(signal_data)
    binary = (signal_data > threshold).astype(int)

    if np.all(binary == 0) or np.all(binary == 1):
        threshold = np.mean(signal_data)
        binary = (signal_data > threshold).astype(int)
        if np.all(binary == 0) or np.all(binary == 1):
            return 0.0

    binary_str = ''.join(binary.astype(str))
    n = len(binary_str)
    complexity = 1
    prefix_len = 1
    i = 0

    while i < n:
        prefix = binary_str[i:i+prefix_len]
        subsequence = binary_str[:i]
        if prefix in subsequence:
            prefix_len += 1
            if i + prefix_len > n:
                break
        else:
            complexity += 1
            i += prefix_len
            prefix_len = 1

    if n > 1:
        max_complexity = n / np.log2(n)
        normalized = complexity / max_complexity
    else:
        normalized = 0.0

    return normalized

# ============================================================================
# NEW TOPOLOGY FUNCTION USING GTDA
# ============================================================================
print("\n" + "=" * 60)
print("Defining new topology metrics with gtda...")
print("=" * 60)

def compute_gtda_metrics(signal_data, embed_dim=3, tau=10, max_points_limit=500):
    """
    Compute persistence entropy from time-delay embedding using gtda.

    Parameters:
    - signal_data: 1D array of EEG signal
    - embed_dim: embedding dimension
    - tau: time delay
    - max_points_limit: maximum number of embedded points to process. Subsamples if exceeded.

    Returns:
    - persistence_entropy: calculated persistence entropy for dimensions 0 and 1
    """

    # Time-delay embedding
    n = len(signal_data)
    n_points_raw = n - (embed_dim - 1) * tau

    if n_points_raw < embed_dim + 1: # Need at least embed_dim + 1 points for a simplex
        return np.nan

    embedded_raw = np.zeros((n_points_raw, embed_dim))
    for i in range(n_points_raw):
        for j in range(embed_dim):
            embedded_raw[i, j] = signal_data[i + j * tau]

    # Subsampling if n_points exceeds limit
    if n_points_raw > max_points_limit:
        # print(f"      Subsampling {n_points_raw} embedded points to {max_points_limit} for gtda calculation.")
        idx = np.random.choice(n_points_raw, max_points_limit, replace=False)
        embedded_points = embedded_raw[idx]
        n_points = max_points_limit
    else:
        embedded_points = embedded_raw
        n_points = n_points_raw

    if n_points < embed_dim + 1: # After subsampling, check again
        return np.nan

    # Reshape embedded_points for gtda (n_samples, n_points, n_dimensions)
    # We are processing one window at a time, so n_samples = 1
    X = embedded_points[None, :, :]

    # Instantiate VietorisRipsPersistence
    # Homology dimensions [0, 1] as requested, excluding 2
    vrp = VietorisRipsPersistence(homology_dimensions=[0, 1], n_jobs=-1)

    # Compute persistence diagrams
    diagrams = vrp.fit_transform(X)

    # Instantiate PersistenceEntropy and compute entropy
    pe = PersistenceEntropy()
    persistence_entropy = pe.fit_transform(diagrams)[0][0] # Returns a 2D array, get the scalar value

    return persistence_entropy

print(" New gtda topology function loaded")

# ============================================================================
# RE-PROCESS DATA WITH NEW METRICS (Using the new gtda function)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 4: Re-loading raw EEG and computing new metrics (gtda enabled)...")
print("=" * 60)

# Load raw data (paths were already identified in previous steps)
awake_file_path = [f for f in !ls /content/ds005620/sub-1010/eeg/*awake*.vhdr if 'acq-EC' in f][0] if !ls /content/ds005620/sub-1010/eeg/*awake*.vhdr else None

raw_sed_files_output = !ls /content/ds005620/sub-1010/eeg/*sed_run*.vhdr
valid_sed_files = [f for f in raw_sed_files_output if os.path.exists(f)]
sed_file_path_candidates = [f for f in valid_sed_files if 'run-1' in f]
sed_file_path = sed_file_path_candidates[0] if sed_file_path_candidates else None

results = []

# TEST_MODE CONFIGURATION (from original notebook cell)
TEST_MODE = True  # Set to False to run full analysis after testing
TEST_WINDOWS_LIMIT = 20 # Number of windows to process in TEST_MODE
MAX_BETTI_POINTS = 500 # Max number of points for Betti calculation (subsamples if exceeded)

states_to_process = [('Awake', awake_file_path)]
if not TEST_MODE:
    if sed_file_path:
        states_to_process.append(('Sedation', sed_file_path))

# Load the existing df from previous steps for original metrics
df = pd.read_csv('/content/eeg_metrics_complete.csv')

for state, filepath in states_to_process:
    if not filepath or not os.path.exists(filepath):
        print(f"Skipping {state}: File path is invalid or does not exist: {filepath}")
        continue

    print(f"\nProcessing {state} state...")

    raw = mne.io.read_raw_brainvision(filepath, preload=True, verbose=False)
    raw.resample(250, verbose=False)

    channel = 'Fp1' if 'Fp1' in raw.ch_names else raw.ch_names[0]
    data = raw.get_data(picks=[channel])[0]

    sfreq = raw.info['sfreq']
    window_samples = int(5.0 * sfreq)
    step_samples = int(1.0 * sfreq)

    n_windows_state = (len(data) - window_samples) // step_samples + 1

    if TEST_MODE:
        n_windows_state = min(n_windows_state, TEST_WINDOWS_LIMIT)

    for i in range(n_windows_state):
        if i % 5 == 0:
            print(f"  Processing window {i}/{n_windows_state} for {state} state...")

        start = i * step_samples
        end = start + window_samples
        window = data[start:end]

        current_df_state = df[df['state'] == state]

        perm_ent = current_df_state.iloc[i]['perm_entropy'] if i < len(current_df_state) else np.nan
        spec_ent = current_df_state.iloc[i]['spectral_entropy'] if i < len(current_df_state) else np.nan
        alpha_pow = current_df_state.iloc[i]['alpha_power'] if i < len(current_df_state) else np.nan

        lz_comp = lempel_ziv_complexity_fixed(window)

        # NEW: gtda persistence entropy
        gtda_persistence_entropy = compute_gtda_metrics(window, max_points_limit=MAX_BETTI_POINTS)

        results.append({
            'state': state,
            'time_sec': i,
            'perm_entropy': perm_ent,
            'spectral_entropy': spec_ent,
            'alpha_power': alpha_pow,
            'lempel_ziv_complexity': lz_comp,
            'gtda_persistence_entropy': gtda_persistence_entropy
        })

    print(f"  Finished processing {n_windows_state} windows for {state} state.")

df_enhanced_gtda = pd.DataFrame(results)

if TEST_MODE:
    print("\n--- TEST MODE RESULTS (gtda Persistence Entropy) ---")
    print(f"Processed {len(df_enhanced_gtda)} windows in TEST_MODE.")

    if not df_enhanced_gtda.empty:
        mean_gtda_pe = df_enhanced_gtda['gtda_persistence_entropy'].mean()
        print(f"Mean gtda_persistence_entropy over {len(df_enhanced_gtda)} test windows: {mean_gtda_pe:.4f}")
        print("First 5 rows of df_enhanced_gtda:")
        print(df_enhanced_gtda.head())
    else:
        print("No data processed in TEST_MODE.")
    print("--------------------------------------------------")
else:
    # Placeholder for full analysis when TEST_MODE is False
    print("Full analysis mode for gtda metrics is not implemented in this step.")
````

````python
import numpy as np
import pandas as pd
import mne
import os
import warnings
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress specific FutureWarning messages from sklearn.utils.deprecation
warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn')

# Import gtda components
from gtda.homology import VietorisRipsPersistence
from gtda.diagrams import PersistenceEntropy

# ============================================================================
# HELPER FUNCTION FOR Lempel-Ziv Complexity (kept from previous steps)
# ============================================================================
def lempel_ziv_complexity_fixed(signal_data):
    """Fixed Lempel-Ziv Complexity"""
    if len(np.unique(signal_data)) == 1:
        return 0.0

    threshold = np.median(signal_data)
    binary = (signal_data > threshold).astype(int)

    if np.all(binary == 0) or np.all(binary == 1):
        threshold = np.mean(signal_data)
        binary = (signal_data > threshold).astype(int)
        if np.all(binary == 0) or np.all(binary == 1):
            return 0.0

    binary_str = ''.join(binary.astype(str))
    n = len(binary_str)
    complexity = 1
    prefix_len = 1
    i = 0

    while i < n:
        prefix = binary_str[i:i+prefix_len]
        subsequence = binary_str[:i]
        if prefix in subsequence:
            prefix_len += 1
            if i + prefix_len > n:
                break
        else:
            complexity += 1
            i += prefix_len
            prefix_len = 1

    if n > 1:
        max_complexity = n / np.log2(n)
        normalized = complexity / max_complexity
    else:
        normalized = 0.0

    return normalized

# ============================================================================
# NEW TOPOLOGY FUNCTION USING GTDA (kept from previous steps)
# ============================================================================
print("\n" + "=" * 60)
print("Defining new topology metrics with gtda...")
print("=" * 60)

def compute_gtda_metrics(signal_data, embed_dim=3, tau=10, max_points_limit=500):
    """
    Compute persistence entropy from time-delay embedding using gtda.

    Parameters:
    - signal_data: 1D array of EEG signal
    - embed_dim: embedding dimension
    - tau: time delay
    - max_points_limit: maximum number of embedded points to process. Subsamples if exceeded.

    Returns:
    - persistence_entropy: calculated persistence entropy for dimensions 0 and 1
    """

    # Time-delay embedding
    n = len(signal_data)
    n_points_raw = n - (embed_dim - 1) * tau

    if n_points_raw < embed_dim + 1: # Need at least embed_dim + 1 points for a simplex
        return np.nan

    embedded_raw = np.zeros((n_points_raw, embed_dim))
    for i in range(n_points_raw):
        for j in range(embed_dim):
            embedded_raw[i, j] = signal_data[i + j * tau]

    # Subsampling if n_points exceeds limit
    if n_points_raw > max_points_limit:
        idx = np.random.choice(n_points_raw, max_points_limit, replace=False)
        embedded_points = embedded_raw[idx]
        n_points = max_points_limit
    else:
        embedded_points = embedded_raw
        n_points = n_points_raw

    if n_points < embed_dim + 1: # After subsampling, check again
        return np.nan

    # Reshape embedded_points for gtda (n_samples, n_points, n_dimensions)
    # We are processing one window at a time, so n_samples = 1
    X = embedded_points[None, :, :]

    # Instantiate VietorisRipsPersistence
    # Homology dimensions [0, 1] as requested, excluding 2
    # Set n_jobs=1 to avoid BrokenProcessPool errors with multiprocessing
    vrp = VietorisRipsPersistence(homology_dimensions=[0, 1], n_jobs=1)

    # Compute persistence diagrams
    diagrams = vrp.fit_transform(X)

    # Instantiate PersistenceEntropy and compute entropy
    pe = PersistenceEntropy()
    persistence_entropy = pe.fit_transform(diagrams)[0][0] # Returns a 2D array, get the scalar value

    return persistence_entropy

print("\u2713 New gtda topology function loaded")

# ============================================================================
# RE-PROCESS DATA WITH NEW METRICS (FULL ANALYSIS)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 4: Re-loading raw EEG and computing new metrics (gtda enabled) - FULL ANALYSIS...")
print("=" * 60)

# --- Revised file path identification using os.walk for robustness ---
target_dir = "/content/ds005620/sub-1010/eeg"
vhdr_files = []
for root, dirs, files in os.walk(target_dir):
    for file in files:
        if file.endswith(".vhdr"):
            vhdr_files.append(os.path.join(root, file))

awake_file_path = None
sed_file_path = None

for file_path in vhdr_files:
    if "task-awake_acq-EC" in file_path and awake_file_path is None:
        awake_file_path = file_path
    elif "task-sed_acq-rest_run-1" in file_path and sed_file_path is None:
        sed_file_path = file_path

print(f"Identified Awake file: {awake_file_path}")
print(f"Identified Sedation file: {sed_file_path}")

# ----------------------------------------------------------------------

results = []

# TEST_MODE CONFIGURATION
TEST_MODE = False  # Set to False for full analysis
# TEST_WINDOWS_LIMIT is no longer used when TEST_MODE is False
MAX_BETTI_POINTS = 500 # Max number of points for Betti calculation (subsamples if exceeded)

states_to_process = []
if awake_file_path and os.path.exists(awake_file_path): # Ensure awake file exists before adding
    states_to_process.append(('Awake', awake_file_path))
if sed_file_path and os.path.exists(sed_file_path): # Ensure sedation file exists before adding
    states_to_process.append(('Sedation', sed_file_path))

# Reload the existing df from previous steps for original metrics
df = pd.read_csv('/content/eeg_metrics_complete.csv')

for state, filepath in states_to_process:
    if not filepath or not os.path.exists(filepath):
        print(f"Skipping {state}: File path is invalid or does not exist: {filepath}")
        continue

    print(f"\nProcessing {state} state...")

    raw = mne.io.read_raw_brainvision(filepath, preload=True, verbose=False)
    raw.resample(250, verbose=False)

    channel = 'Fp1' if 'Fp1' in raw.ch_names else raw.ch_names[0]
    data = raw.get_data(picks=[channel])[0]

    sfreq = raw.info['sfreq']
    window_size_sec = 5
    step_size_sec = 1
    window_samples = int(window_size_sec * sfreq)
    step_samples = int(step_size_sec * sfreq)

    n_windows_state = (len(data) - window_samples) // step_samples + 1

    # Progress printing for full analysis
    print_interval = max(1, n_windows_state // 10) # Print progress about 10 times

    for i in range(n_windows_state):
        if i % print_interval == 0:
            print(f"  Processing window {i}/{n_windows_state} for {state} state...")

        start = i * step_samples
        end = start + window_samples
        window = data[start:end]

        current_df_state = df[df['state'] == state]

        # Safely retrieve existing metrics, will be NaN if index is out of bounds
        perm_ent = current_df_state.iloc[i]['perm_entropy'] if i < len(current_df_state) else np.nan
        spec_ent = current_df_state.iloc[i]['spectral_entropy'] if i < len(current_df_state) else np.nan
        alpha_pow = current_df_state.iloc[i]['alpha_power'] if i < len(current_df_state) else np.nan

        lz_comp = lempel_ziv_complexity_fixed(window)

        # NEW: gtda persistence entropy
        gtda_persistence_entropy = compute_gtda_metrics(window, max_points_limit=MAX_BETTI_POINTS)

        results.append({
            'state': state,
            'time_sec': i,
            'perm_entropy': perm_ent,
            'spectral_entropy': spec_ent,
            'alpha_power': alpha_pow,
            'lempel_ziv_complexity': lz_comp,
            'gtda_persistence_entropy': gtda_persistence_entropy
        })

    print(f"  Finished processing {n_windows_state} windows for {state} state.")

df_enhanced_gtda = pd.DataFrame(results)

print("\n" + "=" * 60)
print("STEP 5: RESULTS SUMMARY (Full Analysis)")
print("=" * 60)

print("\nDataset Overview:")
print(df_enhanced_gtda.groupby('state').size())

print("\nMetrics Summary by State:")
print(df_enhanced_gtda.groupby('state')[['perm_entropy',
                                         'lempel_ziv_complexity', 'alpha_power',
                                         'gtda_persistence_entropy']].mean())

# ============================================================================
# STEP 6: STATISTICAL TESTS (Full Analysis)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 6: Statistical Comparisons (ANOVA)")
print("=" * 60)

metrics_for_anova = ['perm_entropy', 'lempel_ziv_complexity',
                       'alpha_power', 'gtda_persistence_entropy']

for metric in metrics_for_anova:
    awake_vals = df_enhanced_gtda[df_enhanced_gtda['state'] == 'Awake'][metric].dropna()
    sed_vals = df_enhanced_gtda[df_enhanced_gtda['state'] == 'Sedation'][metric].dropna()

    if len(states_to_process) > 1 and len(awake_vals) > 0 and len(sed_vals) > 0:
        f_stat, p_val = f_oneway(awake_vals, sed_vals)
        print(f"{metric:25s}: F={f_stat:8.2f}, p={p_val:.4f}")
    else:
        print(f"Not enough states or data points to perform ANOVA for metric: {metric}")

# ============================================================================
# STEP 7: VISUALIZATIONS (Full Analysis)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 7: Creating visualizations...")
print("=" * 60)

# We now have 4 key metrics to plot: perm_entropy, lempel_ziv_complexity, alpha_power, gtda_persistence_entropy
fig, axes = plt.subplots(2, 2, figsize=(14, 12)) # Adjusted for 4 plots
axes = axes.flatten()

metrics_to_plot_full = ['perm_entropy', 'lempel_ziv_complexity',
                       'alpha_power', 'gtda_persistence_entropy']

if len(states_to_process) > 1: # Only plot if both states are available
    for idx, metric in enumerate(metrics_to_plot_full):
        sns.boxplot(x='state', y=metric, data=df_enhanced_gtda, ax=axes[idx])
        axes[idx].set_title(f'{metric.replace("_", " ").title()}')
        axes[idx].set_xlabel('State')
        axes[idx].set_ylabel(metric.replace("_", " ").title())

    plt.suptitle('Self-Plex Metrics: Awake vs Sedation (Full Analysis)', fontsize=18, y=1.02) # Adjust suptitle position
    plt.tight_layout()
    plt.savefig('self_plex_full_analysis_gtda.png', dpi=300, bbox_inches='tight')
    print("\u2713 Saved: self_plex_full_analysis_gtda.png")
else:
    print("Skipping visualizations: Not enough states to compare.")

# ============================================================================
# STEP 8: SAVE RESULTS (Full Analysis)
# ============================================================================
print("\n" + "=" * 60)
print("STEP 8: Saving enhanced dataset...")
print("=" * 60)

df_enhanced_gtda.to_csv('/content/eeg_metrics_ENHANCED_GTDA.csv', index=False)
print("\u2713 Saved: /content/eeg_metrics_ENHANCED_GTDA.csv")

# ============================================================================
# STEP 9: SELF-PLEX VALIDATION SUMMARY (Full Analysis)
# ============================================================================
print("\n" + "=" * 60)
print("SELF-PLEX FRAMEWORK VALIDATION RESULTS (Full Analysis)")
print("=" * 60)

if len(states_to_process) > 1: # Only summarize if both states are available
    awake_stats = df_enhanced_gtda[df_enhanced_gtda['state'] == 'Awake'][metrics_to_plot_full].mean()
    sed_stats = df_enhanced_gtda[df_enhanced_gtda['state'] == 'Sedation'][metrics_to_plot_full].mean()

    print("\nPREDICTED CHANGES (Self-Plex Framework):")
    print("  Entropy \u2192 DECREASE during sedation")
    print("  Alpha Power \u2192 INCREASE during sedation")
    print("  Topological Complexity (e.g., Persistence Entropy) \u2192 DECREASE during sedation")

    print("\nOBSERVED CHANGES:")
    for metric in metrics_to_plot_full:
        awake_val = awake_stats[metric]
        sed_val = sed_stats[metric]
        if pd.isna(awake_val) or awake_val == 0:
            change = np.nan
            direction = "N/A"
        else:
            change = ((sed_val - awake_val) / awake_val * 100)
            direction = "\u2191" if change > 0 else ("\u2193" if change < 0 else "\u2192")

        print(f"  {metric:25s}: {awake_val:8.4f} \u2192 {sed_val:8.4f} ({change:+6.1f}% {direction})")
else:
    print("Cannot provide validation summary: Not enough states to compare.")

print("\n" + "=" * 60)
print("FULL ANALYSIS COMPLETE! Download the files:")
print("  - eeg_metrics_ENHANCED_GTDA.csv")
if len(states_to_process) > 1:
    print("  - self_plex_full_analysis_gtda.png")
print("=" * 60)
````
